```python
import matplotlib.pyplot as plt 
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

***plotting the data***

```python
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine value")
plt.title("sine wave")
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel("angle")
plt.ylabel("cos value",c="r")
plt.title("Cos Wave")
plt.show()
```

```python
#parabola

x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.title("Parabola")
plt.show()
```

```python
plt.plot(x,y,"r+") # r = red & + = symbol
```

```python
plt.plot(x,y,"g.")
plt.show()
```

```python
plt.plot(x,y,"rx")
plt.show()
```

**mutliple graphs**

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),"r--")
plt.show()
```

**Bar Plot**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,0.9,0.9])
lang = ["Tamil","English","Hindi","Kannada"]
ppl = [100,50,120,2000]
plt.bar(lang,ppl)
plt.xlabel("Language")
plt.ylabel("people")
plt.title("Speaking Languages",c="r")
plt.show()
```

**PieChart**

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,0.5,0.5])
ax.pie(ppl,labels=lang,autopct='%1.1f%%')
plt.show()
```

**Scatter Plot**


```python
x = np.linspace(-5,5,30)
ax = plt.figure().add_axes([0,0,0.5,0.5])
ax.scatter(x,np.cos(x),c='g')
ax.scatter(x,np.sin(x),c='r')
plt.xlabel("angle")
plt.ylabel("sine Value")
plt.title("Sine wave")
plt.show()
```

**3D scatter Plot**

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
